**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
big_salary = data[data['salary'] == ">50K"]
small_salary = data[data['salary'] == "<=50K"]
avg_big = round(big_salary['age'].mean(), 2)
avg_small = round(small_salary['age'].mean(), 2)
sko_big = round(big_salary['age'].std(), 2)
sko_small = round(small_salary['age'].std(), 2)
print ("Average age per people with salary >50K: ", avg_big, "\nStandart deviation: ", sko_big)
print ("Average age per people with salary <=50K: ", avg_small, "\nStandart deviation: ", sko_small)
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
marital_status = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']


men = data[data['sex'] == 'Male']
married_men = men[men['marital-status'].isin(marital_status)]
unmarried_men = men[~men['marital-status'].isin(marital_status)]

married = round(100 * (married_men['salary'] == '>50K').mean(), 2)
unmarried = round(100 * (unmarried_men['salary'] == '>50K').mean(), 2)

print('Part of married', married, "%")
print('Part of unmarried', unmarried, "%")
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
print("Max hours per week:", max_hours)
num_with_max_hours = data[data['hours-per-week'] == max_hours]['hours-per-week'].count()
print("Number of people with max hours-per-week >50K:", num_with_max_hours)
num_with_max_hours_and_salary = data[(data['hours-per-week'] == max_hours) & (data['salary'] == '>50K' )]['hours-per-week'].count()
print("Percentage of working max hours and having salary mote than 50K:", round(100*num_with_max_hours_and_salary/num_with_max_hours, 2), "%")
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
labels = ['young', 'adult', 'retiree']
bins = [16, 35, 70, 100]
data['AgeGroup'] = pd.cut(data['age'], bins=bins, right=False, labels=labels, include_lowest=True)
print(data[['age','AgeGroup']])
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
big_salary_by_group = data[data['salary'] == '>50K'].groupby('AgeGroup',  observed = False).count()

max = big_salary_by_group['age'].max()
print('Amount of people in each age group:\n', big_salary_by_group['age'])

popular_age_group_with_big_salary = big_salary_by_group['age'].idxmax()
print('Name of age group with biggest salary:', popular_age_group_with_big_salary)
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
#group_by_occupation = data.groupby('occupation').count()
#print(group_by_occupation['age'])

def filter_func(group):
    mean_age= group['age'].mean() <= 40
    hours = (group['hours-per-week'] >= 5).all()
    return mean_age & hours

filtered_data = data.groupby('occupation').filter(filter_func)

print("Filtered DataFrame:")
print(filtered_data)
```
