```python
import numpy as np
```

**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
ar = np.random.randint(-10, 10, size=(1,20))
print(ar)
new_arr = np.where((ar>3)&(ar<8), -ar, ar)
print(new_arr)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
ar = np.random.randint(-10, 10, size=(1,20))
print(ar)
max = np.max(ar)
print("Max element:",max)
ar = np.where(ar == max, 0, ar)
print(ar)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
a = np.random.randint(-10, 10, size=(3, 3))
print(a)
b = np.random.randint(-10, 10, size=(3, 3))
print(b)
multiple = np.outer(a, b)
print(multiple)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(-10, 10, size=(8,3))
B = np.random.randint(-10, 10, size=(2,2))
#A = [ [1, 2, 3], [4, 5, 6], [7, 8, 9] ]
#A = np.array(A)
#B = [[1, 2], [1, 3]]
#B = np.array(B)
print("Array A:\n",A)
print("Array B:\n",B)
B=B.flatten()
filteredA = np.isin(A, B)
print(filteredA)
full_line_true = filteredA.all(axis=1)
print("\n:\n", full_line_true)
matching_rows = A[full_line_true]
print("\nСтроки в A, все элементы которых есть в B:\n", matching_rows)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
a = np.random.randint(2, 3, size=(10,3))
print(a)
comp_condition = (a == a[:,[0]])
equal = comp_condition.all(axis=1)
matching_rows = a[~equal]
print("Answer:\n", matching_rows)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
a = np.random.randint(1, 3, size=(10,3))
print(a)
a = np.unique(a, axis=0)
print(a)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
x = [[1,0,1], [2,0,2], [3,0,3], [4,4,4]]
i=0
p=1
while (i<len(x)):
    j=0
    while (j<len(x)):
        if(i<j):
            if(x[i][i]!=0 and i<j):
                p=p*x[i][i]
        j=j+1
    i = i+1

print(p)

y = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
d = np.diag(y)
filteredD = d[d!=0]
p = np.prod(filteredD)

print(p)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
x = np.array([5, 8])
max = None
i = 0
length = len(x)-1
while (i<length):
    if(x[i]==0):
        max=x[i+1]
    i=i+1

if(max!=None):
    while i < length:
        if(x[i] == 0):
            if (x[i+1]>max):
                max = x[i+1]
        i = i+1
    print('Max element after zero:', max)
else:
    print('No element after zero or there is no zero. Max element:', max)

zero_index = np.where(x == 0)[0]
zero_index = zero_index[zero_index<len(x)-1]
after_zero = x[zero_index+1]
if after_zero.size==0:
    print('No element after zero or there is no zero. Max element:', max)
else:
    max = after_zero.max()
    print('Max element after zero:', max)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])
y = [2, 2, 2, 3, 3, 3, 5]

unique = []
uniquec = []
i=0
for i in y:
    if i not in unique:
        unique.append(i)
        count = y.count(i)
        uniquec.append(count)

print('Unique elements:', unique, 'Unique counts:', uniquec)

unique, uniquec = np.unique(x, return_counts=True)
print('Unique elements:', unique, 'Unique counts:', uniquec)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import math

X = np.array([[0,0],[1,1]])
Y = np.array([[1,0],[2,2]])
# X = np.random.rand(1000, 10)
# Y = np.random.rand(1000, 10)

def distance_matrix_python_comp(X, Y):
    return [[math.sqrt(sum((x - y) ** 2 for x, y in zip(a, b))) for b in Y] for a in X]

print(distance_matrix_python_comp(X, Y))

D_numpy = np.sqrt(((X[:, None, :] - Y[None, :, :]) ** 2).sum(axis=2))
print(D_numpy)

from scipy.spatial.distance import cdist

D_scipy = cdist(X, Y, metric='euclidean')
print(D_scipy)

import timeit
t1 = timeit.timeit(lambda: distance_matrix_python_comp(X.tolist(), Y.tolist()), number=1)
t2 = timeit.timeit(lambda: np.sqrt(((X[:, None, :] - Y[None, :, :]) ** 2).sum(axis=2)), number=1)
t3 = timeit.timeit(lambda: cdist(X, Y, metric='euclidean'), number=1)

print(f"Python: {t1*1000:.4f} ms")
print(f"NumPy: {t2*1000:.4f} ms")
print(f"SciPy cdist: {t3*1000:.4f} ms")
```
